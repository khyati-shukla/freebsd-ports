```
import requests
from bs4 import BeautifulSoup
import time

def get_product_details(product_url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(product_url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract ASIN from product URL
        asin = product_url.split("/")[-1]

        # Extract product description
        product_description_elem = soup.find('meta', {'name': 'description'})
        product_description = product_description_elem.get('content').strip() if product_description_elem else 'Not available'

        # Extract manufacturer
        manufacturer_elem = soup.find('a', {'id': 'bylineInfo'})
        manufacturer = manufacturer_elem.text.strip() if manufacturer_elem else 'Not available'

        return asin, product_description, manufacturer
    else:
        print("Failed to retrieve data for URL:", product_url)
        return None, None, None

def scrape_product_data(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        products = soup.find_all('div', {'data-component-type': 's-search-result'})

        for product in products:
            product_url = "https://www.amazon.in" + product.find('a', {'class': 'a-link-normal'})['href']
            product_name = product.find('span', {'class': 'a-size-medium'}).text.strip()
            product_price = product.find('span', {'class': 'a-price'}).find('span', {'class': 'a-offscreen'}).text.strip()

            # Not all products have a rating or number of reviews, so we need to handle that condition.
            rating_elem = product.find('span', {'class': 'a-icon-alt'})
            rating = rating_elem.text.strip() if rating_elem else 'Not available'
            num_reviews_elem = product.find('span', {'class': 'a-size-base'})
            num_reviews = num_reviews_elem.text.strip() if num_reviews_elem else 'Not available'

            print(f"Product Name: {product_name}")
            print(f"Product URL: {product_url}")
            print(f"Product Price: {product_price}")
            print(f"Rating: {rating}")
            print(f"Number of Reviews: {num_reviews}")

            # Get additional details from product URL
            asin, product_description, manufacturer = get_product_details(product_url)
            print(f"ASIN: {asin}")
            print(f"Product Description: {product_description}")
            print(f"Manufacturer: {manufacturer}")

            print('-' * 50)
            # Add a small delay between requests to avoid being blocked
            time.sleep(1)

    else:
        print("Failed to retrieve data. Status code:", response.status_code)

# URL of the product listing page on Amazon.in
base_url = 'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_'
num_pages_to_scrape = 200

for page_num in range(1, num_pages_to_scrape + 1):
    url = base_url + str(page_num)
    print(f"Scraping Page {page_num}...")
    scrape_product_data(url)
```

```
import csv
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import time

# Function to scrape product data from a single listing page
def scrape_listing_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    products = []

    for product in soup.find_all('div', {'data-component-type': 's-search-result'}):
        product_url = product.find('a', {'class': 'a-link-normal'})['href']
        product_name = product.find('span', {'class': 'a-text-normal'}).text.strip()
        product_price = product.find('span', {'class': 'a-offscreen'}).text.strip()
        product_rating = product.find('span', {'class': 'a-icon-alt'})
        rating = product_rating.text.strip() if product_rating else 'N/A'
        num_reviews = product.find('span', {'class': 'a-size-base', 'dir': 'auto'})
        num_reviews = num_reviews.text.strip() if num_reviews else 'N/A'

        products.append({
            'Product URL': urljoin(base_url, product_url),
            'Product Name': product_name,
            'Product Price': product_price,
            'Rating': rating,
            'Number of Reviews': num_reviews,
        })

    return products

# Function to scrape additional product data from a product detail page
def scrape_product_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    product_description = soup.find('meta', {'name': 'description'})
    product_description = product_description['content'] if product_description else 'N/A'

    asin = soup.find('input', {'id': 'ASIN'})
    asin = asin['value'] if asin else 'N/A'

    manufacturer = soup.find('a', {'id': 'bylineInfo'})
    manufacturer = manufacturer['href'].split('/')[1] if manufacturer else 'N/A'

    return {
        'Product URL': url,
        'Description': product_description,
        'ASIN': asin,
        'Product Description': product_description,
        'Manufacturer': manufacturer,
    }

base_url = 'https://www.amazon.in'
search_url = 'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_'

# Number of pages to scrape (at least 20)
num_pages = 20

# Scrape product listings and collect product URLs
product_urls = []
for page_num in range(1, num_pages + 1):
    page_url = search_url + str(page_num)
    response = requests.get(page_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for product in soup.find_all('div', {'data-component-type': 's-search-result'}):
        product_url = product.find('a', {'class': 'a-link-normal'})['href']
        product_urls.append(urljoin(base_url, product_url))

# Scrape additional product data for each product URL
scraped_data = []
for product_url in product_urls[:200]:  # Fetching information from the first 200 URLs
    product_data = scrape_product_page(product_url)
    scraped_data.append(product_data)

# Write the scraped data to a CSV file
csv_filename = 'amazon_bags_data.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['Product Name', 'Product URL', 'Product Price', 'Rating', 'Number of Reviews', 'ASIN', 'Description', 'Product Description', 'Manufacturer']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for data in scraped_data:
        writer.writerow(data)

print(f"Scraping and data export completed. Data saved to '{csv_filename}'.")
```
